```python
#Removing outliers
def IQR_outliers_detect(column):
    q1 = car[column].quantile(.25)
    q3 = car[column].quantile(.75)
    IQR = q3-q1
    print("IQR : {}".format(IQR))
    upper = q3+(1.5*IQR)
    lower = q1-(1.5*IQR)
    print("upper cap : {}".format(upper))
    print("lower cap : {}".format(lower))
```

### Spliting the model based on "KMeans" Clustering, then passing each Cluster though the Linear Regression 

```python
from sklearn.cluster import KMeans
```

```python
distortions=[]
for i in range(1, 15):
    km = KMeans(
        n_clusters=i, random_state=6
    )
    km.fit(car)
    distortions.append(km.inertia_)
```

```python
## C-Distance:
distortions
```

```python
# plot
plt.plot(range(1, 15), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

```python
# Based on the Elbow graph, i am considering 4 clusters (K) for the model building
car_opt = KMeans(n_clusters=3,random_state=6)
car_opt.fit(car)
```

```python
car['Cluster']=car_opt.labels_
```

```python
car.head()
```

```python
car.tail()
```

```python
# Dividing the dataset based on the opt cluster of 3
Cluster_1_car=car[car["Cluster"]==0]
Cluster_2_car=car[car["Cluster"]==1]
Cluster_3_car=car[car["Cluster"]==2]
```

```python
print("Cluster 1:",Cluster_1_car.shape,"\nCluster 2:",Cluster_2_car.shape,"\nCluster 3:",Cluster_3_car.shape)
```

```python
Cluster_1_X=Cluster_1_car.drop(columns=["Cluster","mpg"],axis=1)
Cluster_1_Y=Cluster_1_car["mpg"]
Cluster_2_X=Cluster_2_car.drop(columns=["Cluster","mpg"],axis=1)
Cluster_2_Y=Cluster_2_car["mpg"]
Cluster_3_X=Cluster_3_car.drop(columns=["Cluster","mpg"],axis=1)
Cluster_3_Y=Cluster_3_car["mpg"]
```

```python
from sklearn.linear_model import LinearRegression

car_LiReg1 = LinearRegression()
car_LiReg2 = LinearRegression()
car_LiReg3 = LinearRegression()
```

```python
car_LiReg1.fit(Cluster_1_X,Cluster_1_Y)
car_LiReg2.fit(Cluster_2_X,Cluster_2_Y)
car_LiReg3.fit(Cluster_3_X,Cluster_3_Y)
```

```python
car_LiReg2.coef_
```

```python
print("Below listed the co-efficients for 3 Clusters")
pd.DataFrame([car_LiReg1.coef_,car_LiReg2.coef_,car_LiReg3.coef_],columns=["cyl","disp","hp","wt","acc","yr"]
            ,index=["Coeffients for Cluster 1","Coeffients for Cluster 2","Coeffients for Cluster 3"])
```

```python
print("Score per Cluster\n",
      "\nCluster1 score is:",car_LiReg1.score(Cluster_1_X,Cluster_1_Y),
      "\nCluster2 score is :",car_LiReg2.score(Cluster_2_X,Cluster_2_Y),
      "\nCluster3 score is:",car_LiReg3.score(Cluster_3_X,Cluster_3_Y))
```

### Spliting the model based on Hierarchial Clustering, then passing each Cluster though the Linear Regression 

```python
car=car.drop(columns=['Cluster'])
```

```python
from sklearn.cluster import AgglomerativeClustering 
from scipy.cluster.hierarchy import cophenet, dendrogram, linkage
from scipy.spatial.distance import pdist
model = AgglomerativeClustering(n_clusters=3, affinity='euclidean',  linkage='average')
model.fit(car)
```

```python
car['labels'] = model.labels_
car.head()
```

```python
Cluster_1_data=car[car["labels"]==0]
Cluster_2_data=car[car["labels"]==1]
Cluster_3_data=car[car["labels"]==2]
```

```python
print("Cluster 1 shape is :",Cluster_1_data.shape)
print("\nCluster 2 shape is :",Cluster_2_data.shape)
print("\nCluster 3 shape is:",Cluster_3_data.shape)
```

```python
Cluster_1_X=Cluster_1_data.drop(columns=["labels","mpg"],axis=1)
Cluster_1_Y=Cluster_1_data["mpg"]
Cluster_2_X=Cluster_2_data.drop(columns=["labels","mpg"],axis=1)
Cluster_2_Y=Cluster_2_data["mpg"]
Cluster_3_X=Cluster_3_data.drop(columns=["labels","mpg"],axis=1)
Cluster_3_Y=Cluster_3_data["mpg"]
```

```python
from sklearn.linear_model import LinearRegression

cr_LiReg1 = LinearRegression()
cr_LiReg2 = LinearRegression()
cr_LiReg3 = LinearRegression()
```

```python
cr_LiReg1.fit(Cluster_1_X,Cluster_1_Y)
cr_LiReg2.fit(Cluster_2_X,Cluster_2_Y)
cr_LiReg3.fit(Cluster_3_X,Cluster_3_Y)
```

```python
print("Coefficient of three Clusters are listed below")
pd.DataFrame([cr_LiReg1.coef_,cr_LiReg2.coef_,cr_LiReg3.coef_],columns=["cyl","disp","hp","wt","acc","yr"]
            ,index=["Coeffients for Cluster 1","Coeffients for Cluster 2","Coeffients for Cluster 3"])
```
